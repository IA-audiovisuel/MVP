```python
import joblib
import pandas as pd

rapports=[
    "evaluations_results_mistral-small-3.2-24b-instruct.joblib",
    "evaluations_results_deepseek-v3.1-terminus.joblib",    
    "evaluations_results_mistral-large-2512.joblib",
    "evaluations_results_glm-4.7.joblib",
    "evaluations_results_gpt-oss-120b (copie).joblib",
    "evaluations_results_gpt-oss-120b.joblib"
]

evals=[]
for r in rapports:
    evals+=joblib.load(r)


```

```python
df_evals=pd.DataFrame(evals)
display(df_evals.drop(columns=["question", "question_label", "reponse_reference",  "custom_evaluation_text"]))

metric_cols=["custom_evaluation_score", "factual_correctness (llm based)", "answer_relevancy (llm based)", "answer_accuracy (llm based)"]
grouped= df_evals.groupby(["model", "rag_type", ])[metric_cols].mean().apply(lambda x: round(x, 2)).reset_index()
grouped
```

#### Meileurs modèles par type de RAG 


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

for metric in metric_cols:
    # 2. Créer le grouped bar chart
    plt.figure(figsize=(10, 6))
    sns.barplot(
        data=grouped,
        x='model',
        y=metric,
        hue='rag_type',
        palette='viridis',
        errorbar=None  
    )


    plt.title(f'{metric} par modèle et type de RAG', fontsize=14, fontweight='bold')
    plt.xlabel('Modèle', fontsize=12)
    plt.ylabel('Score d\'évaluation moyen', fontsize=12)
    # plt.ylim(0, 10)  

    # 4. Ajouter les valeurs sur les barres
    for container in plt.gca().containers:
        plt.gca().bar_label(container, fmt='%.1f', padding=3)

    # 5. Ajuster la légende
    plt.legend(title='Type de RAG', title_fontsize=12, fontsize=11)

    plt.tight_layout()
    plt.show()
```

```python
df_activity=pd.read_csv('openrouter_activity_2026-02-03.csv')

print(df_activity.columns)

df_activity.head(2)
```

```python
display(df_activity['user'].unique()[:5])


def get_query_trace(v):
    
    if type(v)!=str:    
        return None

    # check si v contient un nombre suffisant de digits    
    n = len([c for c in v if c.isdigit()])    
    # extraire le hash
    if n >15:
        return v[v.rfind("-")+1:]
    else:
        return None

df_activity["trace_id"]= df_activity["user"].apply(get_query_trace)
df_activity["trace_id"].unique()
```
